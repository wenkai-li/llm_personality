### Test regex map

```python
import re

# Define the sentence
sentence = "Help me complete the sentence with certain Big Five Personality: Openness - high, Conscientiousness - low, Extraversion - low, Agreeableness - median, Neuroticism - high"

# Define the regular expression to find all traits and their descriptors
pattern = r"\b(Openness|Conscientiousness|Extraversion|Agreeableness|Neuroticism)\s*-\s*(low|median|high)"

# Mapping from words to numbers
mapping = {'low': 0, 'median': 1, 'high': 2}

# Search the sentence for matches and replace words with numbers
results = [mapping[match.group(2)] for match in re.finditer(pattern, sentence)]

# Print the results
print(results)
```
