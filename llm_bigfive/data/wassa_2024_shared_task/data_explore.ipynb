```python
trac2 = pd.read_csv("trac2_CONVT_train.csv", on_bad_lines='skip')
trac2.drop(columns=["Emotion", "EmotionalPolarity",	"Empathy", "SelfDisclosure"], inplace=True)
trac2
```

```python
trac3 = pd.read_csv("trac3_EMP_train.csv", on_bad_lines='skip')
trac3.drop(columns=['person_empathy', 'person_distress'], inplace=True)
trac3
```

```python
result_df = (
    trac2.assign(
        person_id=lambda x: x.apply(lambda row: row['person_id_1'] if row['speaker'] == 'Person 1' else row['person_id_2'], axis=1)
    )
    .sort_values(['conversation_id', 'turn_id'])
    .groupby(['conversation_id', 'person_id'])['text']
    .apply(' ||| '.join)
    .reset_index(name='text')
    .groupby('person_id')['text']
    .apply(' ||| '.join)
    .reset_index(name='conversion_history')
)
result_df
```

```python
rich.print(result_df.loc[0, "conversion_history"])
```

```python
trac3.drop(columns=['conversation_id', 'article_id'], inplace=True)
trac3
```

```python
person_ids = trac3['person_id'].unique()
len(person_ids)
```

```python
trac3['person_essay'] = trac3['person_essay'].fillna('')
person_essays = trac3.groupby('person_id')['person_essay'].apply(lambda x: ' ||| '.join(x)).reset_index(name='person_essay')
person_essays
```

```python
merged_df = pd.merge(person_essays, result_df, on='person_id', how='outer')
merged_df
```

```python
missing_positions = merged_df.isnull()
print(missing_positions)
print(merged_df.loc[missing_positions['person_essay'], 'person_essay'])
print(merged_df.loc[missing_positions['conversion_history'], 'conversion_history'])
```
