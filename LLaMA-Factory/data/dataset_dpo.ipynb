```python
import json
from collections import OrderedDict
```

```python
with open("/data/user_data/wenkail/llm_personality/align/data_dpo/train.json") as f:
    train = json.load(f)
```

```python
updated_train = []
for item in train:
    updated_item = {'conversations': item['conversation']}
    for key, value in item.items():
        if key != 'conversation':
            updated_item[key] = value
    updated_train.append(updated_item)
```

```python
with open('/home/wenkail/train.json', 'w') as f:
    json.dump(updated_train, f, indent=4)
```
