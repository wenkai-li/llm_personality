```python
import pandas as pd
import numpy as np
df = pd.read_csv("/home/jiaruil5/personality/llm_personality/psychgenerator_opensource/data/big5_training_data.csv", header = 0, index_col = 0)
df
```

```python
df = df.sample(frac=1, random_state=42)
```

```python
# convert to continuous labels from 0 to 1
from scipy.stats import norm
for i in ['ope_z', 'con_z', 'ext_z', 'agr_z', 'neu_z']:
    df[f'{i}_label'] = norm.cdf(df[i])
df
```

```python
df[['ope_z_label', 'con_z_label', 'ext_z_label', 'agr_z_label', 'neu_z_label']].describe()
```

```python
filter_df = df.drop(columns=["ope_z", "con_z", "ext_z", "agr_z", "neu_z"])
filter_df
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(filter_df, test_size=0.05, random_state=42)
```

```python
train_df
```

```python
train_df, val_df = train_test_split(train_df, test_size=0.1, shuffle=False)
train_df
```
